```python
# v3: 세트별(차량별)로 분류태그 적용된 박스를 사용한 버전
# vehicleWithParcel : 세트 및 차량 정보가 내장된 박스 구분 바코드

from datetime import datetime

case = ''
# vehicle = ''
# parcel = ''
vehicleWithParcel = ''
address = ''
myTimestamp = ''



while(True):
    
    # QR코드 입력 받는다.(무한 반복)
    qr = input("Read a QR code: ")


    # CASE태그가 감지되면 실증CASE태그를 업데이트한다.
    if(qr[0:5]=="CASE-"):
        case = ''
        case = qr
        # parcel = ''
        print(case)

#     # 차량태그가 감지되면 현재 차량태그를 업데이트한다.
#     elif(qr[0:4]=="SET-"):
#         vehicle = ''
#         vehicle = qr
    # parcel = ''
#         print(vehicle)

    # 차량태그가 감지되면 현재 차량태그를 업데이트한다.
    elif(qr[0:4]=="SET-"):
        vehicleWithParcel = ''
        vehicleWithParcel = qr
        print(vehicleWithParcel)
# 
#     # 배송박스태그가 감지되면 현재 배송박스 정보에 저장한다.
#     elif(qr[0:2]=="P-"):
#         parcel = ''
#         parcel = qr
#         print(parcel)

    # 주소태그가 감지되면 배송지 주소태그를 업데이트한다.
    elif(qr[0:5]=="ADDR-"):
        address = ''
        address = qr
        # parcel = ''
        print(address)

    else:
        print("Invalid QR-Code Input!")
        # parcel = ''
        vehicleWithParcel = ''

    # 현재 타임스탬프 값을 생성한다.
    myTimestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    myDate = datetime.now().strftime('%Y-%m-%d')
    # print(qr, s)

    fileName = 'PLZ_FieldTest_' + myDate + '_' + address + '.txt'

    # log string을 생성한다.
    myLog = ''
#     myLog = case + '_' + vehicle  + '_' + parcel  + '_' + address + '_' + myTimestamp
    myLog = case + '_' + vehicleWithParcel + '_' + address + '_' + myTimestamp

    # open the text file with 'append' mode
    with open(fileName, 'a') as f:
        f.write(myLog)
        f.write('\n')
        print("Text save complete!:)
        
    # print whole string
    print(myLog)

```
